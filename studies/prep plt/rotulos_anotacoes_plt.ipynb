```python
fig = plt.figure(figsize=(10,6))

ax = fig.add_subplot(1,1,1)

ax.plot(np.random.randn(1000).cumsum())

ticks = ax.set_xticks([0,250,500,750,1000])

labels = ax.set_xticklabels(['one','two','three','four','five'], rotation=30, fontsize='small')

ax.set_title('My first matplotlib plot')

ax.set_xlabel('Stages')

ax.plot()
```

```python
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt

data = pd.read_csv("https://raw.githubusercontent.com/BrambleXu/pydata-notebook/master/examples/spx.csv", index_col=0, parse_dates=True)

fig = plt.figure(figsize=(12, 6))
ax = fig.add_subplot(1, 1, 1)

spx = data['SPX']
spx.plot(ax=ax, style='g-')

crisis_data = [
    (pd.to_datetime('2007-10-11'), 'Peak of bull market'),
    (pd.to_datetime('2008-03-12'), 'Bear Stearns Fails'),
    (pd.to_datetime('2008-09-15'), 'Lehman Bankruptcy')
]

for date, label in crisis_data:
    ax.annotate(label, xy=(date, spx.loc[date] + 75),
                xytext=(date, spx.loc[date] + 225),
                arrowprops=dict(facecolor='red', headwidth=4, width=2, headlength=4),
                horizontalalignment='left', verticalalignment='top')

ax.set_xlim(pd.Timestamp('2007-01-01'), pd.Timestamp('2011-01-01'))
ax.set_ylim(600, 1800)

ax.set_title('Important dates in the 2008-2009 financial crisis')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)

rect = plt.Rectangle((0.2,0.75), 0.4, 0.15, color='k', alpha=0.3)
circ = plt.Circle((.7, .2),.15, color='b', alpha=0.3)
pgon = plt.Polygon([[.15, .15], [.35, .4], [.2, .6]], color='g', alpha=.5)

ax.add_patch(rect)
ax.add_patch(circ)
ax.add_patch(pgon)
```
