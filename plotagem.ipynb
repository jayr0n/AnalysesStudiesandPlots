# Cores, marcadores e estilos de linha

```python
plt.figure(figsize=(10,6))
plt.plot(np.random.randn(30).cumsum(), 'go--')
```

```python
data = np.random.randn(30).cumsum()

plt.figure(figsize=(8,5))

plt.plot(data, 'k--', label='Default')
plt.plot(data, 'k-', drawstyle='steps-post', label='steps-post')

plt.legend(loc='best')
```

```python
fig = plt.figure(figsize=(10,6))

ax = fig.add_subplot(1,1,1)

ax.plot(np.random.randn(1000).cumsum())

```

```python
fig = plt.figure(figsize=(10,6))

ax = fig.add_subplot(1,1,1)

ax.plot(np.random.randn(1000).cumsum())

ticks = ax.set_xticks([0,250,500,750,1000])

labels = ax.set_xticklabels(['one','two','three','four','five'], rotation=30, fontsize='small')

ax.set_title('My first matplotlib plot')

ax.set_xlabel('Stages')

ax.plot()
```

```python
from numpy.random import randn
fig = plt.figure(figsize=(10,6)); ax=fig.add_subplot(1,1,1)

ax.plot(randn(1000).cumsum(), 'g', label='one')
ax.plot(randn(1000).cumsum(), 'r--', label='two')
ax.plot(randn(1000).cumsum(), 'b.', label='three')

ax.legend(loc='best')
```

# Anotaes e desenhos em uma subplotagem

```python
import pandas as pd
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/BrambleXu/pydata-notebook/master/examples/spx.csv", index_col=0, parse_dates=True)
```

```python
from datetime import datetime

fig = plt.figure(figsize=(14,8)); ax = fig.add_subplot(1,1,1)

data = data
spx = data['SPX']

spx.plot(ax=ax, style='g-')

crisis_data = [
    (datetime(2007,10,11), 'Peak of bull market'),
    (datetime(2008,3,12), 'Bear Stearns Fails'),
    (datetime(2008,9,15), 'Lehman Bankruptcy')]

for date, label in crisis_data:
    ax.annotate(label, xy=(date, spx.asof(date) + 75),
                xytext=(date, spx.asof(date) + 225),
               arrowprops=dict(facecolor='red', headwidth=4, width=2, headlength=4),
               horizontalalignment='left', verticalalignment='top')

ax.set_xlim(['1/1/2007', '1/1/2011'])
ax.set_ylim([600, 1800])

ax.set_title('Important dates in the 2008-2009 financial crisis')
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)

rect = plt.Rectangle((0.2,0.75), 0.4, 0.15, color='k', alpha=0.3)
circ = plt.Circle((.7, .2),.15, color='b', alpha=0.3)
pgon = plt.Polygon([[.15, .15], [.35, .4], [.2, .6]], color='g', alpha=.5)

ax.add_patch(rect)
ax.add_patch(circ)
ax.add_patch(pgon)
```
