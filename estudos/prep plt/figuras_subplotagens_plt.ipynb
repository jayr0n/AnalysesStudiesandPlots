# Figuras e subplotagens

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.arange(10)
plt.plot(data)
```

```python
fig = plt.figure(figsize=(8,4))

ax1 = fig.add_subplot(2,2,1)
_ = ax1.hist(np.random.randn(100), bins=20, color='k', alpha=0.3)

ax2 = fig.add_subplot(2,2,2)
ax2.scatter(np.arange(30), np.arange(30) + 3 * np.random.randn(30))

ax3 = fig.add_subplot(2,2,3)
plt.plot(np.random.randn(50).cumsum(), 'k--')
```

# Cores, marcadores e estilos de linha

```python
plt.figure(figsize=(10,6))
plt.plot(np.random.randn(30).cumsum(), 'go--')
```

```python
data = np.random.randn(30).cumsum()

plt.figure(figsize=(8,5))

plt.plot(data, 'k--', label='Default')
plt.plot(data, 'k-', drawstyle='steps-post', label='steps-post')

plt.legend(loc='best')
```

```python
from numpy.random import randn
fig = plt.figure(figsize=(10,6)); 
ax=fig.add_subplot(1,1,1)

ax.plot(randn(1000).cumsum(), 'g', label='one')
ax.plot(randn(1000).cumsum(), 'r--', label='two')
ax.plot(randn(1000).cumsum(), 'b.', label='three')

ax.legend(loc='best')
```

```python
from numpy.random import randn
fig = plt.figure(figsize=(10,6)); 

ax=fig.add_subplot(2,2,1)
ax.plot(randn(1000).cumsum(), 'g', label='one')
ax=fig.add_subplot(2,2,2)
ax.plot(randn(1000).cumsum(), 'r--', label='two')
ax=fig.add_subplot(2,2,3)
ax.plot(randn(1000).cumsum(), 'b.', label='three')

ax.legend(loc='best')
```
