```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dados = pd.read_excel('C:/Users/Andra/Downloads/dados_vendas.xlsx')
```

```python
dados.head()
```

```python
dados = dados.T.reset_index()
```

```python
dados.columns = dados.iloc[0]
```

```python
dados = dados[1:].drop(index=13).reset_index(drop=True)
```

```python
dados = pd.melt(dados, id_vars=['Month'], var_name=['Produtos'], value_name='Valores')
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.barplot(x='Month', y='Valores', hue='Produtos', data=dados, width=0.8)
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.lineplot(x='Month', y='Valores', hue='Produtos', data=dados)
```
